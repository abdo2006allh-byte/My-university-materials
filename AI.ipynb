```python
class BankAccount:
    def __init__(self , owner , balance):
        self.owner=owner
        self.balance=balance

    def deposit(self, amount):
        self.balance+=amount
        print(f"Deposited ${amount}. New Balance : $ {self.balance}")

    def withdraw(self , amount):
        if amount > self.balance:
            print("Insufficient Funds !")
        else:
            self.balance -=amount
            print(f"Withdraw $ {amount}. Remaining: $ {self.balance}")

my_account = BankAccount("Ahmed" , 1000)
my_account.deposit(500)
my_account.withdraw(2000)
```

```python
class Product:
    def __init__(self , name , price , quantity):
        self.name=name
        self.price=price
        self.quantity=quantity

    def total_price (self):
        return self.price *self.quantity


item1=Product("Laptop", 2000, 2)
print("Total:" , item1.total_price())
```

```python
class Content:
    def __init__(self, title, genre):
        self.title = title
        self.genre = genre
        self.views = 0

    def watch(self):
        self.views += 1

class Movie(Content):
    def __init__(self, title, genre, duration):
        super().__init__(title, genre)
        self.duration = duration

    def get_info(self):
        print(f"Movie: {self.title} | Duration: {self.duration} mins")

class Series(Content):
    def __init__(self, title, genre, episodes):
        super().__init__(title, genre)
        self.episodes = episodes

    def get_info(self):
        print(f"Series: {self.title} | Episodes: {self.episodes}")ر
```
